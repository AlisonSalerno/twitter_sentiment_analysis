```python
tweets_df = tweets[['username','tweet','hashtags']]
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
#using Vader to apply sentiment to each tweet 

analyzer = SentimentIntensityAnalyzer()

tweets_df['label'] = tweets_df['tweet'].map(lambda x: analyzer.polarity_scores(x))
# for sentence in sentences:
#     vs = analyzer.polarity_scores(sentence)
#     print("{:-<65} {}".format(sentence, str(vs)))
```

```python
def comp(d):
    d['compound'] = 0
    return d
```

```python
tweets_df['label']=tweets_df['label'].apply(lambda x : comp(x))
```

```python
def neu(d): 
    if d['neu'] <= .7:
        d['neu'] = 0
    return d
```

```python
tweets_df['label']=tweets_df['label'].apply(lambda x : neu(x))
```

```python
tweets_df['labeled_tweet']=tweets_df['label'].map(lambda x: max(x,key=x.get))
```

```python
tweets_df.groupby('labeled_tweet').count()
```
